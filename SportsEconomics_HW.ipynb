```python
import pandas as pd
from statsmodels.api import OLS , add_constant
from patsy import dmatrices


# 데이터 읽기
csv_file_github = 'https://raw.githubusercontent.com/SeanJSLee/Teaching_YU_DS_SportsEconomics_KR/main/kbo_stat_2023.csv'
df = pd.read_csv( csv_file_github )


variable_list = ['Age', 'CareerYearAt23', 'Height(cm)', 'Weight(kg)',
                    'SigningBonus(100M_KRW,USDtoKRW_1250)',
                    'Salary(100M_KRW,USDtoKRW_1250)', 
                    'AVG', 'G', 'PA', 'AB',
                    'R', 'H', '2B', '3B', 'HR', 'TB', 'RBI', 'SB', 'CS', 'BB', 'HBP',
                    'SO', 'GDP', 'SLG', 'OBP', 'E'
                ]
for variable in variable_list :
    if df.dtypes.to_dict()[variable] == "dtype('object')" :
        df[variable] = pd.to_numeric(df[variable], errors='coerce')
    else :
        continue
    

df['AVG'] = pd.to_numeric(df['AVG'], errors='coerce')
df['SLG'] = pd.to_numeric(df['AVG'], errors='coerce')
df['OBP'] = pd.to_numeric(df['AVG'], errors='coerce')
df = df.dropna(subset=['Salary(100M_KRW,USDtoKRW_1250)'])
df.info()
```

```python
y, X = dmatrices(
            formula_like= 'Salary ~ AVG + G + OBP + E', 
            data        = df.rename(
                                columns={'Salary(100M_KRW,USDtoKRW_1250)':'Salary'}
                                ),
            return_type ='dataframe'
            )

model = OLS(y,X)
results = model.fit()
results.summary()
```
